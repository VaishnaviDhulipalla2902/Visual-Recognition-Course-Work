```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread('girl.jpeg')
plt.imshow(img)
```

```python
#img.shape
rows,cols = img.shape[:2]
```

```python
h = np.zeros(256)
n = rows*cols
img_new = np.zeros(img.shape)
```

```python
for i in range(0,rows):
    for j in range(0,cols):
        h[img[i][j]] = h[img[i][j]] + 1
```

```python
h,total = np.histogram(img.flatten(),256,[0,256])
```

```python
#pdf_list = np.ceil(h*(255)/img.size)
#cdf_list = pdf_list.cumsum()
c = h.cumsum()/n
for i in range(0,rows):
    for j in range(0,cols):
        img_new[i,j] = 255*c[img[i][j]]
```
